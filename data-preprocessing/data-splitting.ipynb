```python
import os
import pandas as pd
import random
from sklearn.utils import shuffle
```

```python
dataset_dir = "word_segmentation_files"
```

```python
neg_dir = dataset_dir +  '/neg'
pos_dir = dataset_dir +  '/pos'
```

```python
neg_files = os.listdir(neg_dir)
pos_files = os.listdir(pos_dir)
```

```python
neg_list = []
# Read files in neg folder 
for fileName in neg_files:
    readDir = neg_dir + '/' + fileName
    neg_list.append(open(file = readDir, mode = 'r', encoding = 'utf-8').readline().strip())
```

```python
pos_list = []
# Read files in pos folder 
for fileName in pos_files:
    readDir = pos_dir + '/' + fileName
    pos_list.append(open(file = readDir, mode = 'r', encoding = 'utf-8').readline().strip())
```

```python
neg_train = len(neg_list) * 80 // 100
print(neg_train)
print(len(neg_list) - neg_train)
```

```python
pos_train = len(pos_list) * 80 // 100
print(pos_train)
print(len(pos_list) - pos_train)
```

```python
print(neg_train + pos_train)
print((len(neg_list) - neg_train) + (len(pos_list) - pos_train))
```
